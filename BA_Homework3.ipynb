```python
#1-List the order number, order date, customer number and customer name (first and last) of orders
# placed in January 2017 sent to Colorado recipients (OrdState).
txt='''
SELECT OrdNo,	OrdDate,	Customer.CustNo,	CustFirstName,	CustLastName
FROM OrderTbl, Customer
WHERE Customer.CustNo=OrderTbl.CustNo AND OrdDate LIKE '2017-01-%' AND OrdState='CO'

'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#4-List the customer number, name (first and last), and balance of Washington
#customers who have placed  one or more orders in February 2017.
txt='''
SELECT Customer.CustNo, CustFirstName,	CustLastName,	CustBal
FROM Customer, OrderTbl
WHERE Customer.CustNo=OrderTbl.CustNo AND CustState='WA' AND OrdDate LIKE '2017-02-%'
GROUP BY Customer.CustNo
HAVING count(OrdNo)>=1
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#6-List product related information if product price is lower than average price
txt='''
SELECT *
FROM Product
WHERE ProdPrice<(SELECT avg(ProdPrice)
                FROM Product)
'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
#7-if product price is lower than average price, find how many times are they ordered?
#List product price as well.
txt='''
SELECT Product.ProdNo, ProdPrice, count(OrdNo) AS 'No of Orders'
FROM Product, OrdLine
WHERE Product.ProdNo=OrdLine.ProdNo AND ProdPrice<(SELECT avg(ProdPrice)
                                                   FROM Product)
GROUP BY Product.ProdNo
'''
Q7 =  psql.sqldf(txt)
Q7
```

```python
#8-List the order number and total amount for orders placed each day.

txt='''
SELECT date(OrdDate) AS 'Date' , count(OrderTbl.OrdNo) AS 'No of Orders', sum(Qty*ProdPrice) AS 'Amount'
FROM OrderTbl, OrdLine, Product
WHERE OrderTbl.OrdNo=OrdLine.OrdNo AND OrdLine.ProdNo=Product.ProdNo
GROUP BY OrdDate
'''
Q8 =  psql.sqldf(txt)
Q8
```

```python
#9-What is the total value of product sold by each employee?
#Included internet orders as well
txt='''
SELECT OrderTbl.EmpNo, sum(Qty*ProdPrice) AS 'Total Amount'
FROM OrderTbl, OrdLine, Product
WHERE OrderTbl.OrdNo=OrdLine.OrdNo AND OrdLine.ProdNo=Product.ProdNo
GROUP BY OrderTbl.EmpNo
'''
Q9=  psql.sqldf(txt)
Q9
```

```python
#10-List the total commission earned by each employee in each month of 2017.
txt='''
SELECT OrderTbl.EmpNo, strftime('%m', date(OrdDate)) AS Month, sum(Qty*ProdPrice*EmpCommRate) AS 'Comission Earned'
FROM Employee, OrderTbl, OrdLine, Product
WHERE Employee.EmpNo=OrderTbl.EmpNo AND OrderTbl.OrdNo=OrdLine.OrdNo AND OrdLine.ProdNo=Product.ProdNo AND OrdDate LIKE '2017-%'
GROUP BY OrderTbl.EmpNo, strftime('%m', date(OrdDate))
'''
Q10 =  psql.sqldf(txt)
Q10
```

```python
#11-List the customer first and last name and order number and employee first name and last name
#who handled that order where order is given in February 2017.
#.If order is internet order, include them as well
#Note:If order is internet order, employee first name and last name will be null
txt='''
SELECT CustFirstName, CustLastName, OrdNo, EmpFirstName, EmpLastName
FROM Customer LEFT JOIN OrderTbl ON Customer.CustNo=OrderTbl.CustNo
              LEFT JOIN Employee ON OrderTbl.EmpNo=Employee.EmpNo
WHERE  OrdDate LIKE'2017-02-%'
'''
Q11=  psql.sqldf(txt)
Q11
```

```python
#12-List the total commission earned by each employee. Includeemployee first and last name,
#supervisor last name and first name if any.If employee does not have supervisor, supervisor last name and first name will be null.
txt='''
SELECT OrderTbl.EmpNo,	Employee.EmpFirstName,	Employee.EmpLastName, sum(Qty*ProdPrice*Employee.EmpCommRate) AS 'Comission Earned',
      Employee_1.EmpFirstName AS 'Supervisor First Name', Employee_1.EmpLastName	AS 'Supervisor Last Name'
FROM Product LEFT JOIN OrdLine ON Product.ProdNo=OrdLine.ProdNo
             LEFT JOIN OrderTbl ON OrderTbl.OrdNo=OrdLine.OrdNo
             INNER JOIN Employee ON Employee.EmpNo=OrderTbl.EmpNo
             LEFT JOIN Employee_1 ON Employee.SupEmpNo=Employee_1.EmpNo
GROUP BY OrderTbl.EmpNo

'''
Q12 =  psql.sqldf(txt)
Q12
```

```python
#14-How many times each product are ordered? What is the total value of orders taken by each product?
#Whatis the total quantity ordered for each product?
txt='''
SELECT Product.ProdNo, count(OrdNo) AS 'No of Orders', sum(Qty) AS 'Quantity Ordered', sum(Qty*ProdPrice) AS 'Total Value'
FROM Product LEFT JOIN OrdLine ON Product.ProdNo=OrdLine.ProdNo
GROUP BY Product.ProdNo
'''
Q14 =  psql.sqldf(txt)
Q14
```
