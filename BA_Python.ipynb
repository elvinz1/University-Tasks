```python
#Q4-Write a method to calculate age for a given date of birth year.
#Use this method on orer to calculate age of person if date of birth is 1944
# if age is the name of function, print(age(1944))
from datetime import date
def calculationofage(birth_year):
    today=date.today()
    age=int(today.year) - birth_year
    return age
print(calculationofage(1944))
```

```python
#Q5- write a for loop or method  to calculate summation of first n numbers
#i.e if n=3, it will print or return 6 (1+2+3)
n=int(input('Sum of how many numbers: '))
sum=0
for num in range(1, n + 1, 1):
    sum = sum + num
print("Sum of first ", n, "numbers is: ", sum)
```

```python
#Q12-List Percentage of  the Married
import pandas as pd

data = pd.read_csv(
    'https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv')

df = pd.DataFrame(data)

value_counts = df['Married'].value_counts()

percentages = value_counts.apply(lambda x: x / len(df) )

print(percentages)
```

```python
#14-Get the mean vale of people's Income if region is south and age is below 40
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv')

filtered_data = df[(df['Region'] == 'South') & (df['Age'] < 40)]

mean_income = filtered_data['Income'].mean()

print(mean_income)
```

```python
#Q15-Create new column named IncomebyLimit by dividing Income to Limit.
#and get descriptive statistics of IncomebyLimit.
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv')

df['IncomebyLimit'] = df['Income'] / df['Limit']

stats = df['IncomebyLimit'].describe()

print(stats)
```
