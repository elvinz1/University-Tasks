```python
!pip install pandasql
import pandas as pd
import pandasql as psql
Customer=pd.read_excel("https://github.com/ogut77/BA/blob/main/Customer.xlsx?raw=true")
Employee=pd.read_excel("https://github.com/ogut77/BA/blob/main/Employee.xlsx?raw=true")
Employee_1=pd.read_excel("https://github.com/ogut77/BA/blob/main/Employee.xlsx?raw=true")
Product=pd.read_excel("https://github.com/ogut77/BA/blob/main/Product.xlsx?raw=true")
OrdLine=pd.read_excel("https://github.com/ogut77/BA/blob/main/OrdLine.xlsx?raw=true")
OrderTbl=pd.read_excel("https://github.com/ogut77/BA/blob/main/OrderTbl.xlsx?raw=true")
```

```python
#1-List Employee first name and last name from Employee Table
txt='''
SELECT EmpFirstName,	EmpLastName
FROM Employee
'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#2-List all customer information from Customer Table where Customer city is Seattle   or
#Customer State is Colorada ('CO')
txt='''
SELECT *
FROM Customer
WHERE CustCity='Seattle' OR CustState='CO'
'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#3-How many Customer are there in Customer Table where Customer Balance is 0.
txt='''
SELECT count(*) AS 'Number of Customer with 0 Balance'
FROM Customer
WHERE CustBal=0
'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#4-List product with the name Color in it ?
txt='''
SELECT *
FROM Product
WHERE ProdName LIKE '%Color%'
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#5-Find average, minimum, maximum price of  products?How many product are there in Product Table?
txt='''
SELECT min(ProdPrice) AS 'MinPrice', max(ProdPrice) AS 'MaxPrice', avg(ProdPrice) AS 'AvgPrice', count(distinct ProdNo) AS 'NoOfProduct'
FROM Product

'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#6-What is the total value of product (ProdQOH*ProdPrice) at the inventory for each product manafuacturer?
txt='''
SELECT ProdMfg, ProdQOH*ProdPrice AS 'Total Value of Product at Hand'
FROM Product
GROUP BY ProdMfg
'''
Q6=  psql.sqldf(txt)
Q6
```

```python
#7-How many different products are there in each order?.
txt='''
SELECT OrdNo, count(distinct ProdNo) AS 'No of Product'
FROM OrdLine
GROUP BY OrdNo
'''
Q7 =  psql.sqldf(txt)
Q7
```

```python
#8-List order with more than 2 products  in it.
txt='''
SELECT OrdNo, count(distinct ProdNo) AS 'No of Product'
FROM OrdLine
GROUP BY OrdNo
HAVING count(distinct ProdNo)>2
'''
Q8 =  psql.sqldf(txt)
Q8
```

```python
#9-How many times each product are ordered?
txt='''
SELECT ProdNo, count(OrdNo) as 'No of Orders'
FROM OrdLine
GROUP BY ProdNo
'''
Q9 =  psql.sqldf(txt)
Q9
```

```python
#10-Combine Names (CustFirstName and CustLastName) form Customer table
#with neames from Employee Table (EmpFirstName and EmpLastName )
txt='''
SELECT CustFirstName AS 'First Name', CustLastName AS 'Last Name'
FROM Customer
UNION
SELECT EmpFirstName, EmpLastName
FROM Employee

'''
Q10 =  psql.sqldf(txt)
Q10
```

```python
#11-Find name (First and last name) that both appers in Employee and Customer Table
txt='''
SELECT CustFirstName AS 'First Name', CustLastName AS 'Last Name'
FROM Customer
INTERSECT
SELECT EmpFirstName, EmpLastName
FROM Employee
'''
Q11 =  psql.sqldf(txt)
Q11
```

```python
#12-Find name (First aand last name) that appears in Employee but not  Customer Table
txt='''
SELECT EmpFirstName AS 'First Name', EmpLastName AS 'Last Name'
FROM Employee
EXCEPT
SELECT CustFirstName, CustLastName
FROM Customer
'''
Q12 =  psql.sqldf(txt)
Q12
```

```python
#13-Find name (First aand last name) that appears in Customer but not  Employee Table
txt='''
SELECT CustFirstName AS 'First Name', CustLastName AS 'Last Name'
FROM Customer
EXCEPT
SELECT EmpFirstName, EmpLastName
FROM Employee
'''
Q13 =  psql.sqldf(txt)
Q13
```

```python
#14. List the average balance and number of customers by city. Only include
#customers residing in Washington State (WA).
txt='''
SELECT CustCity, avg(CustBal) AS 'Avg of Customer Balance', count(distinct CustNo) AS 'No of Customers'
FROM Customer
WHERE CustState='WA'
GROUP BY CustCity
'''
Q14 =  psql.sqldf(txt)
Q14
```

```python
#15. How many product are there without ProdNextShipDate.
#What is their average price?
txt='''
SELECT count(distinct ProdNo) AS 'No of Products', avg(ProdPrice) AS 'Average Price'
FROM Product
WHERE ProdNextShipDate is NULL
'''
Q15 =  psql.sqldf(txt)
Q15
```
